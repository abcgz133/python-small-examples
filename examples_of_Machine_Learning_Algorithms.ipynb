```python
%matplotlib inline
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


# 创建 DataFrame
data = {
    'Location_Chicago': [0, 0, 0, 1, 0],
    'Merchant Category_Grocery': [0, 0, 0, 0, 0],
    'Location_New York': [0, 0, 0, 1, 0],
    'Customer Gender_Female': [1, 1, 0, 0, 1],
    'target': ['Negative', 'Negative', 'Positive', 'Negative', 'Negative']
}

df = pd.DataFrame(data)

# 分离特征和目标变量
X = df.drop(columns='target')
y = df['target']

# 特征名称
feature_names = X.columns.tolist()
target_name = y.unique().tolist()
# 创建并训练决策树分类器
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# 绘制决策树
plt.figure(figsize=(12, 8))
plot_tree(clf,
          feature_names=feature_names,
          class_names = target_name,
          filled=True,
          rounded=True,
          fontsize=10)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
from sklearn.base import clone
from collections import Counter
import matplotlib.pyplot as plt

# 创建假设的数据集
data = {
    'Transaction Amount': [50.0, 200.0, 75.0, 150.0, 30.0],
    'Merchant Category': ['Grocery', 'Electronics', 'Restaurant', 'Clothing', 'Pharmacy'],
    'Transaction Time': [14.5, 21.0, 19.75, 18.5, 16.0],
    'Location': ['New York', 'Los Angeles', 'Chicago', 'San Francisco', 'Boston'],
    'Card Type': ['Credit', 'Debit', 'Credit', 'Credit', 'Debit'],
    'Customer Age': [35, 45, 28, 32, 25],
    'Customer Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Is Fraudulent': [0, 1, 0, 0, 1]
}

df = pd.DataFrame(data)

# 将分类变量转换为数值型（例如：One-Hot 编码）
df_encoded = pd.get_dummies(df, columns=['Merchant Category', 'Location', 'Card Type', 'Customer Gender'])

# 分离特征和目标变量
X = df_encoded.drop('Is Fraudulent', axis=1)
y = df_encoded['Is Fraudulent']

# 打印原始数据集（部分）
print("原始数据集（部分）：")
print(df.head())
print("\n编码后的特征数据：")
print(X.head())

# 定义一个函数来随机选择部分特征
def random_feature_subset(X, max_features='sqrt'):
    """
    随机选择部分特征。

    参数:
    X (DataFrame): 特征数据。
    max_features (str or float): 选择特征的数量，'sqrt' 表示平方根，默认值。

    返回:
    list: 选中的特征索引。
    """
    if isinstance(max_features, str) and max_features == 'sqrt':
        max_features = int(np.sqrt(X.shape[1]))
    elif isinstance(max_features, float):
        max_features = int(max_features * X.shape[1])

    features = np.random.choice(X.shape[1], max_features, replace=False)
    return features

# 构建每棵树
n_estimators = 5  # 减少树的数量以简化演示
trees = []

for i in range(n_estimators):
    print(f"\n\n程序本身的开始。现在开始构建第 {i+1} 棵树：")

    # 有放回抽样生成子样本集
    subset = resample(df_encoded, replace=True, n_samples=len(df_encoded), random_state=42 + i)
    X_subset = subset.drop('Is Fraudulent', axis=1)
    y_subset = subset['Is Fraudulent']

    print("子样本集大小：", len(subset))
    print("子样本集：")
    print(subset)

    # 随机选择一部分特征
    selected_features = random_feature_subset(X_subset, max_features='sqrt')
    X_subset_selected = X_subset.iloc[:, selected_features]
    print("开始随机选择一部分特征")
    print("选中的特征索引：", selected_features)
    print("选中的特征名称：", X_subset.columns[selected_features].tolist())
    print("子样本集中选中的特征数据：")
    print(X_subset_selected)
    print("\n- y_subset -")
    print(y_subset)

    # 构建决策树（注意这里使用了克隆以确保每次都是独立的树）
    print("开始构建特征树")
    # 决策树默认使用： GINI不纯度分类
    tree_decision = clone(DecisionTreeClassifier(max_depth=3, random_state=42 + i))
    tree_decision.fit(X_subset_selected, y_subset)
    tree.plot_tree(tree_decision,
                   feature_names=X_subset_selected.columns.tolist(),
                   fontsize=9)
    plt.show()
    # 记录哪些特征被选中以及对应的树
    trees.append((tree_decision, selected_features))

print("\n所有树构建完成。\n")

def predict_with_random_forest(X_test, trees):
    """
    使用随机森林进行预测。

    参数:
    X_test (DataFrame): 测试数据。
    trees (list of tuples): 决策树及其对应的特征索引列表。

    返回:
    list: 最终预测结果。
    """
    predictions = []

    for i in range(len(X_test)):
        votes = []

        for tree, selected_features in trees:
            # 只使用选定的特征进行预测
            X_test_selected = X_test.iloc[i, selected_features].values.reshape(1, -1)
            prediction = tree.predict(X_test_selected)
            votes.append(prediction[0])
            print(f"第 {i+1} 个测试样本，使用树 {trees.index((tree, selected_features)) + 1} 的预测结果：{prediction[0]}")

        # 对每个样本进行投票
        most_common = Counter(votes).most_common(1)[0][0]
        predictions.append(most_common)
        print(f"第 {i+1} 个测试样本的最终预测结果：{most_common}")

    return predictions

# 测试数据（假设是前两行）
X_test = X.iloc[:2, :]
print("\n测试数据：")
print(X_test)

# 进行预测
final_predictions = predict_with_random_forest(X_test, trees)
print("\n最终预测结果汇总：", final_predictions)

# 总结输出
print("\n总结输出：")

# 打印原始数据集（部分）
print("原始数据集（部分）：")
print(df.head())

# 打印编码后的特征数据
print("\n编码后的特征数据：")
print(X.head())

# 打印测试数据
print("\n测试数据：")
print(X_test)

# 打印最终预测结果
print("\n最终预测结果汇总：", final_predictions)

```
